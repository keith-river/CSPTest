## 系統的探索法と発見的探索法
|系統的探索法 | 発見的探索法|
| :--- | :--- |
| 発見的知識を用いない探索|発見的知識を用いた探索 |
| 深さ優先探索 | 山登り法 |
| 幅優先探索 | 最適探索 |
| 反復深化探索 | 最良優先探索 |
|  | A*（Aスター）アルゴリズム |


```python
#  教科書 pp.20～23   図3.2,  図3.3, 図3.4  
# 木構造，グラフ構造，問題設定，データ設定

ABC='_ABCDEFGHIJKLMNOPQRSTUVWXYZ'

NN=range(1,27)
Nodes={ i:ABC[i] for i in NN }
def N(n):
    return Nodes[n]

def alpha(nodes):
    return [ N(n) for n in nodes ]

def one_hops(node):
    return [ e[1] for e in Edges if e[0]==node ]

Edges = [ (1,2),(1,3),(1,4),(2,5),(2,6),(3,7),(3,8),(4,9),(4,10),(6,11),(6,12) ]
A_edges = [ (N(e[0]),N(e[1])) for e in Edges ]

Node_levels = [0,0,1,1,1,2,2,2,2,2,2,3,3]
```

```python
#  教科書 pp.20～23   図3.2,  図3.3, 図3.4  
#  深さ優先探索 Depth First Search

OpenList=[]    # Jobs to be done  未決，まだ片付いてないもの
ClosedList=[]  # done  既決，片付いたもの

def trace(node):  # 状態モニタリング用
    print(f'candidate = {N(node)}') 
    print(f'OpenList= {alpha(OpenList)} | ClosedList = {alpha(ClosedList)}')
    print(f'node={N(node)} -> candidate={alpha(candidate(node))}')  
    print()
          
def candidate(node): # nodeから1ホップで辿れて， OpenListにもClosedListにも入っていないもの
    return [ e[1] for e in Edges if e[0]==node and (not e[1] in ClosedList and not e[1] in OpenList) ]
        
def DFS(ini,fin):
    '''深さ優先探索（Depth First Search）, Stack反復利用（非再帰版)'''
    global OpenList,ClosedList
# step 1.
    OpenList,ClosedList=[ini],[]
    while True:
# step 2.
        if OpenList==[]: return None
# step 3.
        cand = OpenList.pop(0)
        ClosedList.append(cand)
        trace(cand)
# step 4.       
        if cand==fin: return cand
# step 5,6
        nexts = candidate(cand)
        if nexts != []:
            OpenList = nexts + OpenList  # OpenList+=nexts  # BFS
            
    return None            

ini,fin = 1,12
print(N( DFS(ini,fin) ))
```

```python
#  教科書 pp.20～23   図3.2,  図3.3, 図3.4  
#  幅優先探索 Breadth First Search

OpenList=[]    # Jobs to be done  未決，まだ片付いてないもの
ClosedList=[]  # done  既決，片付いたもの

def trace(node):  # 状態モニタリング用
    print(f'candidate = {N(node)}') 
    print(f'OpenList= {alpha(OpenList)} | ClosedList = {alpha(ClosedList)}')
    print(f'node={N(node)} -> candidate={alpha(candidate(node))}')  
    print()
          
def candidate(node): # nodeから1ホップで辿れて， OpenListにもClosedListにも入っていないもの
    return [ e[1] for e in Edges if e[0]==node and (not e[1] in ClosedList and not e[1] in OpenList) ]
        
def BFS(ini,fin):
    '''幅優先探索（Breadth First Search）, Queue反復利用'''
    global OpenList,ClosedList
# step 1.
    OpenList,ClosedList=[ini],[]
    while True:
# step 2.        
        if OpenList==[]:
            return None
# step 3.        
        cand = OpenList.pop(0)
        ClosedList.append(cand)
        trace(cand)
# step 4.        
        if cand==fin:
            return cand
# step 5. 6.        
        nexts = candidate(cand)        
        if nexts != []:
            OpenList+=nexts  # BFS
    return None            

ini,fin = 1,12
print(N( BFS(ini,fin) ))
```

```python
#  教科書 pp.20～23   図3.2,  図3.3, 図3.4 
# 反復深化探索 Iterative Deepening Depth First Search
    
def candidate(node): # nodeから1ホップで辿れて， OpenListにもClosedListにも入っていないもの
    return [ e[1] for e in Edges if e[0]==node and (not e[1] in ClosedList and not e[1] in OpenList) ]

def trace(node): # 状態モニタリング
    print(f'candidate = {N(node)}@level_{Node_levels[node]}') 
    print(f'OpenList= {alpha(OpenList)}  |  ClosedList = {alpha(ClosedList)}')
    print(f'node={N(node)} | 1 hop candidates = {alpha(candidate(node))}')  
    print() 

def ID_DFS(lim,ini,fin):
    '''反復深化探索 Iterative Deepening Depth First Search'''
    global OpenList,ClosedList
# step 2
    print(f'----- Iterative level #{lim} -----') 
    OpenList,ClosedList = [ini],[]
    while True:
# step 3    
        if OpenList==[]: return None
# step 4    
        cand = OpenList.pop(0)
        ClosedList.append(cand)
        trace(cand)
# step 5    
        if cand==fin: return cand
# step 6     
        if Node_levels[cand]<lim:
            nexts = candidate(cand)
            if nexts != []:
                OpenList = nexts + OpenList  # OpenList += nexts  # BFS            
    return None            

ini,fin = 1,12

for lim in range(1,5):
    ret = ID_DFS(lim,ini,fin)
    if ret:
        print(f'Found: {fin} == {ret}')
    else:
        print(f'Not found')
```

```python
import queue
stack=queue.LifoQueue()

def push(stack,x):
    stack.put(x)
    
def pop(stack):
    return stack.get()

operators=[(push,5),(push,8),(push,10),(push,19),(pop,0),(pop,0),(push,7)]

for o in operators:
    if o[0]==push:
        push(stack,o[1])
    else:
        pop(stack)

def show(stack):
    while not stack.empty():
        x=pop(stack)
        print(x)
    
show(stack)
```

```python
# Binary search tree <= create, search
class Tree():
    def __init__(self,v):
        self.value=v
        self.left=None # NULL
        self.right=None # NULL
        print(f'new node:{v}')
            
    def in_oder(self):
        print(f'{self.value}')
        if self.left: self.left.in_oder()
        if self.right: self.right.in_oder()

    def pre_oder(self):
        if self.left: self.left.pre_oder()
        print(f'{self.value}')
        if self.right: self.right.pre_oder()

    def post_oder(self):
        if self.left: self.left.post_oder()
        if self.right: self.right.post_oder()
        print(f'{self.value}')

    def search(self,x):
        print(f'search {x} to {self.value}')
        if self.value==x:
            print(f'Found {x} @ {self.value}')
            return self
        elif x < self.value:
            if self.left!=None:
                print(f'{x} -> left of {self.value}')
                return self.left.search(x) 
            else: return None
        else:
            if self.right!=None:
                print(f'{x} -> right of {self.value}')
                return self.right.search(x) 
            else: return None
            
    def add(self,x):
        if x<self.value:
            print(f'Add({x}) as Left child of {self.value}')
            if self.left==None:
                self.left=Tree(x)
            else:
                self.left.add(x)
        else:
            print(f'Add({x}) as Right child of {self.value}')    
            if self.right==None:
                self.right=Tree(x)
            else:
                self.right.add(x)
        return self

    def is_leaf(self):
        return self.left==self.right==None

    def min(self):
        node = self.left
        while node.left:
           node = node.left
        return node.value

    def max(self):
        node = self.right
        while node.right:
           node = node.right
        return node.value
          
    def left_max(self,node):
        cand = node.left
        while cand.right != None:
            cand = cand.right
        return cand.value
    def right_min(self,node):
        cand = node.right
        while cand.left != None:
            cand = cand.left
        return cand.value

    def remove(self,parent,x):
      if parent is None: return None
      if x < parent.value:
        parent.left = self.remove(parent.left,x)
      elif x > parent.value:
        parent.right = self.remove(parent.right,x) 
      else:  #  x == parent.value
        if parent.is_leaf():  #  delete the node(leaf) itself
            parent=None
        elif parent.left==None or parent.right==None:
            parent = parent.left or parent.right  # delete the node, set parent child its grand-child
        else: # in case of removing node that has two child-node
            parent.value=parent.left_max(parent)  # or right_min(parent)
            parent.left=parent.left.remove(parent.left,parent.value) # or parent.right.remove()
      return parent

#  basic operations test for binary tree
Node=[35,62,74,9,77,13,23,3,53,8]
print('Binary Tree creation test:')  
BTree=Tree(Node[0])
STree=BTree

for node in Node[1:]:
    BTree.add(node)
    print()

BTree.in_oder()
print()
BTree.pre_oder()
print()
BTree.post_oder()
print()

print(f'Minimum node = { BTree.min() }')
print(f'Maximum node = { BTree.max() }')
print()

print('Binary Tree Search test:')  
x=53
target = BTree.search(x)
if target!=None:
    print(target.value)
else:
    print(f'Not Found {x}')

BTree.remove(BTree,62)
BTree.in_oder()
```
